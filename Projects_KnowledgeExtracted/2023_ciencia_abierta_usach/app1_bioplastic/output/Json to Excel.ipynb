## Grouo of Json to Excel

```python
df_master = pd.DataFrame()
for file in json_list_name:
    print(file)
    data = open(file)
    json_file = json.load(data)
    
    metadata = json_file['metadata']
    film_development = json_file['data_extracted']['film_development']
    materials = json_file['data_extracted']['materials']
    df_conc_proper = pd.DataFrame.from_dict(json_file['data_extracted']['concentration_and_properties'], orient="columns")
    df_data_extract = pd.concat([df_conc_proper], keys=[materials[0]], names=['materials'])
    df_data_extract = pd.concat([df_data_extract], keys=[film_development[0]], names=['film_development'])
    df_data_extract = pd.concat([df_data_extract], keys=[metadata['doi']], names=['doi'])
    df_data_extract = pd.concat([df_data_extract], keys=[metadata['date_doc']], names=['date_doc'])
    df_data_extract = pd.concat([df_data_extract], keys=[metadata['authors']], names=['authors'])
    df_data_extract = pd.concat([df_data_extract], keys=[metadata['title']], names=['title'])
    df_data_extract = pd.concat([df_data_extract], keys=[metadata['type_doc']], names=['type_doc'])
    df_data_extract = pd.concat([df_data_extract], keys=[metadata['name_file']], names=['name_file'])
    df_master = pd.concat([df_master, df_data_extract], axis=1)
df_master
```

```python
df_save = df_master.fillna(0)
df_save
```

```python
df_save.to_excel('..\..\extraction_biopolymers_data.xlsx')
```
